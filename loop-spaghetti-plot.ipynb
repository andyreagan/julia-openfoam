```julia
function plotEnsembleSpag(directory_suffix)
    folder = "/Users/andyreagan/work/2014/2014-11foamLab-julia/"
    forecastFlux = readcsv("$(folder)results/forecastFlux$(directory_suffix).csv");
    println(size(forecastFlux))
    # this is the same...
    # analysisFlux = readcsv("results/analysisFlux-$(dec(trial,3)).csv");
    truthFlux = readcsv("$(folder)results/truthFlux$(directory_suffix).csv");
    println(size(truthFlux))
    # make the spaghetti plot
    fig = PyPlot.matplotlib[:pyplot][:figure](figsize=(10,5))
    ax = fig[:add_axes]([0.2,0.2,0.7,0.7])
    ax[:plot](0:10:10*length(truthFlux)-1,truthFlux,"b",label="Hidden Truth")
    for i=1:Nens
        # println(size(forecastFlux[i,:]))
        # only label the first one
        if i<2
            ax[:plot](0:10:10*length(forecastFlux[i,:])-1,forecastFlux[i,:]',"r",label="Ensemble Members")
        else
            ax[:plot](0:10:10*length(forecastFlux[i,:])-1,forecastFlux[i,:]',"r")
        end
    end
    ax[:set_title]("$(directory_suffix)")
    ax[:set_ylabel]("Flux \$\\phi\$",size=18)
    ax[:set_xlabel]("Time \$t\$",size=18)
    # ax[:set_xlabel]("time")
    # ax[:set_ylabel]("unscaled 9o'clock face flux avg")

    PyPlot.matplotlib[:pyplot][:savefig]("$(folder)loop-spaghetti$(directory_suffix).png")
end
```
